```python
# 导入包
import csv
import jieba

# 读取csv文件
csv_reader = csv.reader(open("C:/Users/11040/Desktop/weibonlp-master/data/combined_csv.csv"))


jieba.load_userdict("SogouLabDic.txt")
jieba.load_userdict("dict_baidu_utf8.txt")
jieba.load_userdict("dict_pangu.txt")
jieba.load_userdict("dict_sougou_utf8.txt")
jieba.load_userdict("dict_tencent_utf8.txt")
jieba.load_userdict("my_dict.txt")

stopwords = {}.fromkeys([ line.rstrip() for line in open('C:/Users/11040/Desktop/weibonlp-master/data/stopwords.txt') ])

sentences = []

# 循环读出每一行进行处理
i = 1 
for row in csv_reader:
    
    # 评论内容用结巴分词以空格分词
    comments = jieba.cut(row[3]) 
    
    for i in comments:
        if i not in stopwords:
            sentences.append(i)
            
fo = open("C:/Users/11040/Desktop/weibonlp-master/data/0219/keyword2.dat", "a+")
        #fo = open("/Users/kimmeen/Downloads/P_Weibo/%s"%user_id, "w")
    
for j in sentences:
          
           fo.write(j)
           fo.write(' ')
     
fo.write('\n')
fo.close()
        

print(sentences)
```

```python
from jieba import analyse

tfidf = analyse.extract_tags

for line in open("C:/Users/11040/Desktop/weibonlp-master/data/0219/keyword.dat"):
    
    text = line

    keywords = tfidf(text,allowPOS=('a','an'))

    result=[]

    for keyword in keywords:
        
        result.append(keyword)

    #print(result)
    fo = open("C:/Users/11040/Desktop/weibonlp-master/data/0219/adj.dat", "a+")
    
    for j in result:
          
        fo.write(j)
        fo.write(' ')
     
    fo.write('\n')
    fo.close()

print("关键词提取完毕！")
```
